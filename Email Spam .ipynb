```python
import numpy as np
import os
from collections import Counter
```

```python
folder='Email/'
```

```python
files=os.listdir(folder)
len(files)
```

```python
Emails=[folder + file for file in files]
len(Emails)
```

```python
word=[]
for Email in Emails:
    f=open(Email,encoding='latin-1')
    blob=f.read()
    word=word+blob.split(" ")
        
```

```python
len(word)
```

```python
for i in range(len(word)):
    if not word[i].isalpha():
        word[i]=""
```

```python
word_dict=Counter(word)
len(word_dict)
```

```python
del word_dict[""]
```

```python
word_dict=word_dict.most_common(3000)
word_dict
```

```python
for i in word_dict:
    print(i[0])
```

```python
features=[]
labels=[] 

for Email in Emails:
    f=open(Email,encoding='latin-1')
    blob=f.read().split()
    data=[]
    
    for i in word_dict:
        data.append(blob.count(i[0]))
    features.append(data)
    
    if 'spam' in Email:
        labels.append(1)
    if 'ham' in Email :
        labels.append(0)
```

```python
len(features)
```

```python
len(labels)
```

```python
features=np.array(features)
features
```

```python
labels=np.array(labels)
labels
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,labels,test_size=0.2,random_state=7)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
classifier.fit(X_train,y_train)
```

```python
 new_email="Hi you have won a lottery of 10 lacs. to get your reward kindly share your phone number."
```

```python
#new_email="hii there"
```

```python
input_mail=[]
for i in word_dict:
    input_mail.append(new_email.split(" ").count(i[0]))
```

```python
input_mail=np.array(input_mail)
input_mail
```

```python
classifier.predict(input_mail.reshape(1,3000))
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
